VB.*> matches any verb using the . as a wildcard and the * quantifier to match 0 or more occurrences of any character.
<DT>?<JJ>*<NN> matches any noun phrase <RB.?> matches any adverb using the . as a wildcard and the optional quantifier
to match 0 or 1 occurrence of any character. ? is an optional quantifier, matching either 0 or 1 adverbs
